```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
vuelos = pd.read_csv('vuelos.csv')
vuelos
```

```python
vuelos.describe()
```

```python
vuelos['Precio_boleto'].hist(bins=50)
plt.xlabel('Precio del boleto')
plt.ylabel('Cantidad de boletos')
plt.title('Distribución de precios de boletos')
plt.show()
```

```python
vuelos.groupby('aerolinea')['dep_delay'].mean().sort_values().plot(kind='bar')
plt.ylabel('Retraso promedio (minutos)')
plt.title('Retraso promedio por aerolínea')
plt.show()
```
